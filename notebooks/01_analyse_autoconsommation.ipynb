```python
df_clean=df.copy()
```

```python
region_mapping={
    11:"Île-de-France",
    24:"Centre-Val-de-Loire",
    27:"Bourgogne-Franche-Comté",
    28:"Normandie",
    32:"Hauts-de-France",
    44:"Grand-Est",
    52:"Pays-de-la-Loire",
    53:"Bretagne",
    75:"Nouvelle-Aquitaine",
    76:"Occitanie",
    84:"Auvergne-Rhônes-Alpes",
    93:"Provence-Alpes-Côte-d'Azur"
}
df_clean["nom_region_clean"]=df_clean["code_region"].map(region_mapping)
```

```python
df_clean.groupby("code_region")["nom_region_clean"].unique()
```

```python
df_clean["nom_region_clean"].unique()
```

```python
df_clean["operation_valeur_manquantes"]=df_clean["nombre_doperations_actives"].isna()
```

```python
df_clean["nombre_doperations_actives_clean"]=(
    df_clean["nombre_doperations_actives"]
    .fillna(0)
    .astype(int)
)
```

```python
df_clean["trimestre_date"] = pd.to_datetime(
    df_clean["trimestre"],
    errors="coerce"
)
```

```python
df_clean[["trimestre", "trimestre_date"]].head(10)
```

```python
df_clean["nom_region_clean"]
```

```python
df_clean["année"]= df_clean["trimestre_date"].dt.year
```

```python
df_clean["trimestre_date"].isna()
```

```python
df_clean["nombre_doperations_actives_clean"].isna().sum()
```

```python
df_clean[df_clean["nombre_doperations_actives"].isna()]
```

```python
df_clean.columns
```

```python
df_clean=df_clean.rename(columns={"année":"annee"})
```

```python
df_clean.columns
```

```python
df_powerbi=df_clean[
    ["trimestre_date",
     "annee",
     "code_region",
     "nom_region_clean",
     "nombre_doperations_actives_clean",
     "operation_valeur_manquantes"
    ]
    ].copy()
```

```python
df_powerbi.head()
```

```python
df_powerbi=df_powerbi.rename(columns={
    "nom_region_clean":"nom_region",
    "nombre_doperations_actives_clean":"operations_actives",
    "operation_valeur_manquantes":"valeur_operation_manquante"
})
```

```python
df_powerbi.head()
```

```python
df_powerbi.info()
```

```python
df_powerbi.to_csv(
    "../data/processed/autoconsommation_region_clean.csv",
    index=False,
    encoding ="utf-8-sig"
)
```
